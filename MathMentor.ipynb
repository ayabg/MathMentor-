```python
from transformers import AutoTokenizer, AutoModelForCausalLM

# Charger un modèle de base disponible (GPT-2 comme exemple)
tokenizer = AutoTokenizer.from_pretrained("gpt2")
model = AutoModelForCausalLM.from_pretrained("gpt2")
```

```python
def chatbot_response(prompt):
    # Générer une réponse basée sur le modèle
    inputs = tokenizer(prompt, return_tensors="pt")
    outputs = model.generate(inputs.input_ids, max_length=50, num_return_sequences=1, do_sample=True)
    response = tokenizer.decode(outputs[0], skip_special_tokens=True)
    return response

# Tester une réponse
user_input = "Explique comment résoudre x^2 - 4 = 0"
print(chatbot_response(user_input))
```
